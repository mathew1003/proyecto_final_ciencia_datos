# 04 · Conclusiones

El volumen manda. Los minutos (MP) y el volumen de tiro (FGA y 3PA) explican la mayor parte de la variación de PTS. La eficiencia (FG%, 3P%, 2P%, FT%) ayuda a afinar, pero no reemplaza al volumen.

Eras distintas, hábitos distintos. En la era 2010s+ se ve un aumento claro en 3PA/FGA: más triples y una dispersión mayor en la anotación por jugador.

Posición vs. rol real. La variable Pos aporta, pero segmentar por perfil de tiro (especialmente 3PA/FGA) distingue mejor a los anotadores que la etiqueta tradicional PG/SG/SF/PF/C.

Edad y ciclo de rendimiento. La relación con PTS tiende a un pico en la segunda mitad de los 20s; luego se estabiliza o cae ligeramente al controlar por minutos y uso.

Baseline interpretable que cumple. El modelo lineal con imputación + escalado + One-Hot para Pos entrega un desempeño razonable (ver reports/metrics.json). Es simple de explicar y rápido de entrenar.

Dónde se equivoca más. Los errores son mayores en jugadores con pocos minutos o perfiles muy atípicos (varianza alta / menos datos representativos).

Lo que aprendimos del ranking. FGA, MP y 3PA aparecen arriba en feature_ranking.csv. La eficiencia entra como ajuste fino: mejora, pero no lidera por sí sola.

Sin fuga de información. Todas las variables del modelo provienen de la misma temporada, y el split por jugador evita contaminar train/test con el mismo sujeto.

Comunicación clara. El dashboard en Streamlit permite revisar KPIs, distribuciones, relaciones y métricas sin abrir cuadernos, lo cual facilita la defensa y la verificación.

Limitaciones y mejoras

Limitaciones: no modela interacciones (p. ej., MP×FGA), no incorpora métricas de uso/ritmo del equipo ni efectos no lineales por segmentos finos.

Mejoras sugeridas: probar Ridge/Lasso para estabilizar coeficientes, modelos no lineales (Random Forest / XGBoost), interacciones y segmentación por era/posición; además, validación más estricta (k-fold por jugador) y calibración de errores para bajo minutaje.

Cierre corto

Con 8 variables fijas y datos de la misma temporada, se puede explicar y predecir PTS con un error razonable y una narrativa clara: jugar más y tirar más pesa más que solo “tirar mejor”; la eficiencia suma, pero el volumen sigue siendo el motor. El dashboard deja todo a la vista para evaluarlo sin fricción.
